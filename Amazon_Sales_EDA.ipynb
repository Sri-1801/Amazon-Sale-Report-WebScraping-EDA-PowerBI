```python
# Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Excel file
df = pd.read_excel("Amazon Sale Report.csv.xlsx")

# Display first few rows to confirm
df.head()
```

```python
df.isnull().sum()
```

```python
# Fill missing rating_count values with 0 (or you can use mean/median too)
df['rating_count'] = df['rating_count'].fillna(0)
```

```python
df.isnull().sum()
```

```python
# Check for duplicate rows
df.duplicated().sum()
```

```python
# Clean price columns (handles messy symbols like â‚¹ etc.)
df['discounted_price'] = df['discounted_price'].replace(r'[^0-9.]', '', regex=True).astype(float)
df['actual_price'] = df['actual_price'].replace(r'[^0-9.]', '', regex=True).astype(float)
```

```python
df['discount_percentage'] = ((df['actual_price'] - df['discounted_price']) / df['actual_price']) * 100
```

```python
df[['discounted_price', 'actual_price', 'discount_percentage']].head()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
sns.histplot(df['discounted_price'], bins=30, kde=True, color='skyblue')
plt.title('Distribution of Discounted Prices')

plt.subplot(1,2,2)
sns.histplot(df['discount_percentage'], bins=30, kde=True, color='salmon')
plt.title('Distribution of Discount Percentages')

plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='discount_percentage', y='rating_count', alpha=0.6, color='teal')
plt.title('Discount % vs Rating Count')
plt.xlabel('Discount Percentage')
plt.ylabel('Number of Ratings')
plt.show()
```

```python
category_summary = df.groupby('category').agg({
    'discounted_price': 'mean',
    'discount_percentage': 'mean',
    'rating_count': 'mean'
}).sort_values(by='rating_count', ascending=False)

print(category_summary.head(10))
```

```python
import pandas as pd

# Load the Amazon Sale Report file
df = pd.read_excel("Amazon Sale Report.csv.xlsx")

# Check if it loaded correctly
df.head()


```

```python
# Clean up weird characters like â‚¹, commas, ₹ etc.
df['discounted_price'] = (
    df['discounted_price']
    .astype(str)
    .str.replace(r'[^\d.]', '', regex=True)  # keep only digits and dot
    .replace('', '0')  # replace empty with 0
    .astype(float)
)

df['actual_price'] = (
    df['actual_price']
    .astype(str)
    .str.replace(r'[^\d.]', '', regex=True)
    .replace('', '0')
    .astype(float)
)

df['discount_percentage'] = (
    df['discount_percentage']
    .astype(str)
    .str.replace(r'[^\d.]', '', regex=True)
    .replace('', '0')
    .astype(float)
)

df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')
```

```python
category_summary = df.groupby('category').agg({
    'discounted_price': 'mean',
    'discount_percentage': 'mean',
    'rating_count': 'mean'
}).sort_values('rating_count', ascending=False)
```

```python
df[['discounted_price', 'actual_price', 'discount_percentage', 'rating_count']].head()
```

```python
category_summary = df.groupby('category').agg({
    'discounted_price': 'mean',
    'discount_percentage': 'mean',
    'rating_count': 'mean'
}).sort_values('rating_count', ascending=False)

category_summary.head(10)

```

```python

import matplotlib.pyplot as plt
import seaborn as sns

top_categories = category_summary.head(10)

plt.figure(figsize=(10,6))
sns.barplot(y=top_categories.index, x=top_categories['rating_count'], palette='viridis')
plt.title('Top 10 Categories by Average Rating Count', fontsize=14)
plt.xlabel('Average Rating Count')
plt.ylabel('Category')
plt.show()

```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='discount_percentage', y='rating_count', alpha=0.6)
plt.title('Discount Percentage vs Rating Count', fontsize=14)
plt.xlabel('Discount Percentage')
plt.ylabel('Rating Count')
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df, x='discounted_price', y='rating_count', alpha=0.6, color='orange')
plt.title('Discounted Price vs Rating Count', fontsize=14)
plt.xlabel('Discounted Price (₹)')
plt.ylabel('Rating Count')
plt.show()
```
